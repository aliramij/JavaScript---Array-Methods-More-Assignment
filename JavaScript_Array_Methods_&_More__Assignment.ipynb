```python
# Question 6 : Write code using the reduce method to calculate the total value of all items
# in a shopping cart. Each item has a price and quantity. Also find the most expensive
# single item in the cart.
# // Shopping cart with items
# const cart = [
# { name: "Laptop", price: 50000, quantity: 1 },
# { name: "Mouse", price: 500, quantity: 2 },
# { name: "Keyboard", price: 1500, quantity: 1 },
# { name: "Monitor", price: 15000, quantity: 1 }
# ];


'''
const total_value_of_items = cart.reduce((prev,curr)=> {
  return prev + (curr.price * curr.quantity)
},0)
console.log(total_value_of_items)
const mostExpensive = cart.reduce((maxItem, currentItem) => {
  return maxItem.price > currentItem.price ? maxItem : currentItem;
}, 0);
console.log(`Most expensive item is ${mostExpensive.name} which price is ${mostExpensive.price}`)

'''
```

```python
# Question 9 : Write a function called calculateTotal that can accept any number of
# prices as arguments and returns their sum. Use the rest operator to handle the variable
# number of parameters.

'''

function calculateTotal (...price) {
  return price.reduce((prev,curr)=> {
    return prev + curr
  },0)
}
console.log(calculateTotal(100,500,250,900,800))
'''
```
